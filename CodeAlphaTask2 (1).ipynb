##EXPLORATORY DATA ANALYSIS(EDA)

##1.CHECKING THE DATATYPES AND COLUMNS IN DATASET

```python
import pandas as pd
data=pd.read_csv('/content/ElectricCarData_Clean (1).csv')
print(data.info())
```

##2.CHECKING WHETHER THE DATA HAS THE NULL VALUES AND THEIR COUNT


```python
print(data.isna())
print(data.isna().sum())
```

##3.REMOVING THE NULL VALUES AND CHECKING THE COUNT AGAIN

```python
new_data = data.dropna(axis=1)
print(new_data.isna().sum())
```

##4.CHECKING THE DUPLICATE ENTRIES


```python
print(new_data.duplicated())
```

##5.COUNTING FREQUENCY OF EACH UNIQUE VALUE


```python
print(data['Brand'].value_counts())
```
